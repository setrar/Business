```julia
using Plots
```

```julia
quarters = ["Q1","Q2", "Q3", "Q4", "Q5", "Q6", "Q7", "Q8", "Q9"]
# total_revenue = [1149, 10080, 23940, 33152, 51576, 79212, 18684, 16606]
total_cogs         = [0, 441+0, 371+485, 301+813, 740+738, 5228+1904, 7712+0, 6560+1091, 20508+10589]
total_gross_margin = [0, 1149,    4078,    6454,    6702,     10872,  11008,     18684,       23288]

# gross_profit = [2160, 7680, 19140, 31680, 48991, 75241, 110242, 153994]
opex = [0, 500, 1000, 2000, 2479, 3072, 3807, 4719, 4719]
net_income = [-1374, -1022, 639, 2101, 2576, 4778, 4636, 7598, 10176]; # net_earnings
```

```julia
total_revenue = [
    (margin + cogs) for (margin, cogs) in zip(total_gross_margin, total_cogs)
];
```

1. Profitability Ratios

$$
\text{Gross Profit Margin} = \left(\frac{\text{Total Revenue} - \text{COGS}}{\text{Total Revenue}}\right) \times 100\%
$$


```julia
gross_profit_margin = [
    (((revenue - cogs) / revenue) * 100) for (revenue, cogs) in zip(total_revenue, total_cogs)
];
```

```julia
current_assets      = [18106, 19964, 23264, 26674, 35419, 40687, 54287, 66422]
current_liabilities = [  504,  1837,  3100,  3867,  7753,  8414, 14811, 16697];
```

$$
\text{Net Profit Margin} = \left(\frac{\text{Net Profit}}{\text{Total Revenue}}\right) \times 100\%
$$


```julia
net_profit_margin = [
    ((profit / revenue) * 100) for (revenue, profit) in zip(total_revenue, net_income)
];
```

```julia
# Plotting the gross profit margin
p1 = plot(gross_profit_margin
    , label = "Gross Profit Margin ((revenue - cogs) / revenue)"
    , linewidth=2, marker=:square
)
p1 = plot!(net_profit_margin
    , label = "Net Profit Margin (profit / revenue)"
    , linewidth=2, marker=:diamond
)
plot(p1
    , legend=true
    # , title="Key Financial Indicators"
    , xlabel="Q Periods", ylabel="(%) In Percentage Points "
    , size = (800,600)
)
```

$$
\text{Current Ratio} = \frac{\text{Current Assets}}{\text{Current Liabilities}}
$$


```julia
current_ratio = [
    ((asset / liability) * 100) for (asset, liability) in zip(current_assets, current_liabilities)
];
```
